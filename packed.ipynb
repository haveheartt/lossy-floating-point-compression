```python
import numpy as np
import struct
import os
import matplotlib.pyplot as plt

# Generate samples from uniform, Gaussian, and exponential distributions
def generate_distributions(n_samples=100000):
    uniform = np.random.uniform(-100, 100, n_samples)
    gaussian = np.random.normal(0, 10, n_samples)
    exponential = np.random.exponential(10, n_samples)
    
    return {
        'uniform': uniform,
        'gaussian': gaussian,
        'exponential': exponential
    }

class FloatCompressor:
    # Total number of mantissa bits in IEEE 754 single precision 
    NMANTISSA = 23
    # Number of exponent bits in IEEE 754 single precision
    NEXPONENT = 8
    # Total bits in IEEE 754
    TOTAL_BITS = 32

    def __init__(self, mantissa_bits):
        self.mantissa_bits = mantissa_bits
        self.mantissa_bitmask = 0

        # Adjust the received bit number to some reasonable value:
        if self.mantissa_bits < 5:
            self.mantissa_bits = 5
        if self.mantissa_bits > self.NMANTISSA:
            self.mantissa_bits = self.NMANTISSA

        # Fill up the lower N bits
        for i in range(self.NMANTISSA - self.mantissa_bits):
            self.mantissa_bitmask |= (0x1 << i)
        # And now negate it to get the correct mask:
        self.mantissa_bitmask = ~self.mantissa_bitmask

        # Set the Magic numbers
        if self.mantissa_bits == self.NMANTISSA:
            self.rounding = 0
        else:
            self.rounding = 0x1 << (32 - (1 + 8 + self.mantissa_bits) - 1)

        self.vmax = 0x7f7 << 20
        self.vmax |= 0x000fffff ^ self.rounding
        
        # Calculate total bits in compressed representation
        self.total_compressed_bits = 1 + self.NEXPONENT + self.mantissa_bits  # sign + exponent + reduced mantissa
        
        # Calculate bytes needed for compressed format
        self.bytes_needed = (self.total_compressed_bits + 7) // 8  # Ceiling division to get bytes
        
        print(f"Compressing from 32 bits to {self.total_compressed_bits} bits ({self.bytes_needed} bytes)")

    def reduce_float_precision(self, value):
        # Check if any compression is to be made
        if self.mantissa_bits == self.NMANTISSA:
            return value

        # Check for NaN, etc
        if not np.isfinite(value):
            return value

        # Convert float to its 32-bit integer representation
        ivalue = np.float32(value).view(np.int32)

        # Safety-check if value (omitting the sign-bit) is lower than vmax (avoid overflow)
        if (ivalue & 0x7fffffff) < self.vmax:
            ivalue += self.rounding

        # Do the compression
        ivalue &= self.mantissa_bitmask

        # Convert the integer back to a float
        fvalue = ivalue.view(np.float32)
        return fvalue.item()  # Convert numpy float32 to Python float
        
    def compress_to_bytes(self, value):
        # First reduce precision
        reduced = self.reduce_float_precision(value)
        
        # Get the binary representation
        bits = np.float32(reduced).view(np.int32)
        
        # Extract the sign bit
        sign = (bits >> 31) & 0x1
        
        # Extract the exponent
        exponent = (bits >> 23) & 0xFF
        
        # Extract the mantissa
        mantissa = (bits & 0x7FFFFF) >> (self.NMANTISSA - self.mantissa_bits)
        
        # Pack into a compact representation
        packed = (sign << (self.NEXPONENT + self.mantissa_bits)) | (exponent << self.mantissa_bits) | mantissa
        
        # Convert NumPy int32 to Python int, then to bytes
        packed_int = int(packed)
        result = packed_int.to_bytes(self.bytes_needed, byteorder='little')
        return result
    
    def decompress_from_bytes(self, byte_data):
        # Convert bytes to int
        packed = int.from_bytes(byte_data, byteorder='little')
        
        # Extract the sign bit, exponent, and mantissa
        sign = (packed >> (self.NEXPONENT + self.mantissa_bits)) & 0x1
        exponent = (packed >> self.mantissa_bits) & 0xFF
        mantissa = packed & ((1 << self.mantissa_bits) - 1)
        
        # Reconstruct the full binary representation
        full_mantissa = mantissa << (self.NMANTISSA - self.mantissa_bits)
        
        # Build the full 32-bit representation
        bits = (sign << 31) | (exponent << 23) | full_mantissa
        
        # Convert back to float
        return np.uint32(bits).view(np.float32).item()
    
def save_to_binary(data, filename, compress=False, compressor=None):
    output_dir = 'output'
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    base_filename = os.path.basename(filename)
    
    file_path = os.path.join(output_dir, base_filename)
    
    if compress and compressor:
        # Open the file in binary write mode
        with open(file_path, 'wb') as f:
            # Write each compressed float to the file
            for value in data:
                compressed_bytes = compressor.compress_to_bytes(value)
                f.write(compressed_bytes)
    else:
        # Save the full precision data
        data.astype('float32').tofile(file_path)
    
def load_from_binary(filename, decompress=False, compressor=None):
    file_path = os.path.join('output', filename)
    
    if decompress and compressor:
        # Calculate the number of floats in the file
        file_size = os.path.getsize(file_path)
        num_floats = file_size // compressor.bytes_needed
        
        # Read and decompress the data
        decompressed_data = np.zeros(num_floats, dtype=np.float32)
        with open(file_path, 'rb') as f:
            for i in range(num_floats):
                byte_data = f.read(compressor.bytes_needed)
                if not byte_data:
                    break
                decompressed_data[i] = compressor.decompress_from_bytes(byte_data)
        
        return decompressed_data
    else:
        # Load the full precision data
        return np.fromfile(file_path, dtype=np.float32)

def get_file_size(filename):
    return os.path.getsize(os.path.join('output', filename))

def compare_statistics(original, compressed, name):
    print(f"\nStatistics for {name}:")
    print(f"Original Mean: {np.mean(original)}, Compressed Mean: {np.mean(compressed)}")
    print(f"Original Variance: {np.var(original)}, Compressed Variance: {np.var(compressed)}")
    print(f"Original Std Dev: {np.std(original)}, Compressed Std Dev: {np.std(compressed)}")

def compute_mse(original, compressed):
    return np.mean((original - compressed) ** 2)

def plot_distributions(original, compressed, name):
    plt.figure(figsize=(12, 6))
    plt.hist(original, bins=100, alpha=0.5, label='Original')
    plt.hist(compressed, bins=100, alpha=0.5, label='Compressed')
    plt.title(f'{name} Distribution')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

# Generate a relevantly large number of floating-point numbers from three distributions
data = generate_distributions()

# Lossy compression function to zero out between 8 and 16 least significant bits of the mantissa.
compressor = FloatCompressor(mantissa_bits=12) # Adjust mantissa_bits as needed

# Save full-precision data to binary files
save_to_binary(data['uniform'], 'uniform_original.bin')
save_to_binary(data['gaussian'], 'gaussian_original.bin')
save_to_binary(data['exponential'], 'exponential_original.bin')

# Save the bit-packed data to binary files
save_to_binary(data['uniform'], 'uniform_compressed.bin', compress=True, compressor=compressor)
save_to_binary(data['gaussian'], 'gaussian_compressed.bin', compress=True, compressor=compressor)
save_to_binary(data['exponential'], 'exponential_compressed.bin', compress=True, compressor=compressor)

# Measure and compare on-disk file sizes
orig_uniform_size = get_file_size('uniform_original.bin')
comp_uniform_size = get_file_size('uniform_compressed.bin')
orig_gaussian_size = get_file_size('gaussian_original.bin')
comp_gaussian_size = get_file_size('gaussian_compressed.bin')
orig_exponential_size = get_file_size('exponential_original.bin')
comp_exponential_size = get_file_size('exponential_compressed.bin')

print(f"Uniform Original Size: {orig_uniform_size} bytes")
print(f"Uniform Compressed Size: {comp_uniform_size} bytes")
print(f"Compression Ratio: {orig_uniform_size / comp_uniform_size:.2f}x")

print(f"Gaussian Original Size: {orig_gaussian_size} bytes")
print(f"Gaussian Compressed Size: {comp_gaussian_size} bytes")
print(f"Compression Ratio: {orig_gaussian_size / comp_gaussian_size:.2f}x")

print(f"Exponential Original Size: {orig_exponential_size} bytes")
print(f"Exponential Compressed Size: {comp_exponential_size} bytes")
print(f"Compression Ratio: {orig_exponential_size / comp_exponential_size:.2f}x")

# Load and decompress the data for verification
decomp_uniform = load_from_binary('uniform_compressed.bin', decompress=True, compressor=compressor)
decomp_gaussian = load_from_binary('gaussian_compressed.bin', decompress=True, compressor=compressor)
decomp_exponential = load_from_binary('exponential_compressed.bin', decompress=True, compressor=compressor)

# Compare statistical parameters of the distributions before and after compression
compare_statistics(data['uniform'], decomp_uniform, "Uniform")
compare_statistics(data['gaussian'], decomp_gaussian, "Gaussian")
compare_statistics(data['exponential'], decomp_exponential, "Exponential")

# Compute Mean Squared Error between original and compressed data
print(f"\nMSE for Uniform: {compute_mse(data['uniform'], decomp_uniform)}")
print(f"MSE for Gaussian: {compute_mse(data['gaussian'], decomp_gaussian)}")
print(f"MSE for Exponential: {compute_mse(data['exponential'], decomp_exponential)}")

# Plot original and compressed data to observe distribution changes
plot_distributions(data['uniform'], decomp_uniform, "Uniform")
plot_distributions(data['gaussian'], decomp_gaussian, "Gaussian")
plot_distributions(data['exponential'], decomp_exponential, "Exponential")
```
